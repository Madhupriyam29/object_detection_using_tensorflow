```python
import torch
from PIL import Image
import matplotlib.pyplot as plt
from google.colab import files

# Upload an image
uploaded = files.upload()  # Upload the image file
img = Image.open(next(iter(uploaded)))  # Open the uploaded image (get the first uploaded file)

# Load the pre-trained YOLOv5 model (small version for speed)
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')  # 'yolov5s' is the smaller version

# Perform inference (object detection)
results = model(img)

# Show the results (image with bounding boxes)
results.show()  # This will display the image with bounding boxes drawn

# Get the detected object names and their confidence scores
detections = results.pandas().xywh[0]  # Convert results to pandas dataframe

# Print the structure of detections to see what columns are available
print(detections.head())

# Print the detected objects and their confidence scores (adjusting column names if necessary)
print(detections[['name', 'confidence']])  # Only print name and confidence
```
